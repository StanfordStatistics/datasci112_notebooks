# Types of Joins

Let's compare baby names in 1920 and 2020. To do this, we have to join two data sets together.

```python
import pandas as pd
data_dir = "http://dlsun.github.io/pods/data/names/"
df_1920 = pd.read_csv(data_dir + "yob1920.txt", header=None,
                      names=["Name", "Sex", "Count"])
df_2020 = pd.read_csv(data_dir + "yob2020.txt", header=None,
                      names=["Name", "Sex", "Count"])
```

```python
df_joined = df_1920.merge(df_2020, on=["Name", "Sex"])
df_joined
```

Let's look for the name Maya in the joined data.

```python
df_joined[df_joined["Name"] == "Maya"]
```

It is there in the 2020 data...

```python
df_2020[df_2020["Name"] == "Maya"]
```

...but not in the 1920 data.

```python
df_1920[df_1920["Name"] == "Maya"]
```

In order to appear in the joined data, a key must be present in _both_ tables.

## Other Kinds of Joins

By default, Pandas does an **inner join**.

But we can also do an **outer join**, **left join**, or **right join**.

```python
df_joined_outer = df_1920.merge(df_2020, on=["Name", "Sex"],
                                how="outer")
df_joined_outer[df_joined_outer["Name"] == "Maya"]
```

```python
df_joined_right = df_1920.merge(df_2020, on=["Name", "Sex"],
                                how="right")
df_joined_right[df_joined_right["Name"] == "Maya"]
```

## Many-to-Many Joins

Why does Mary appear 4 times in the following join?

```python
df_1920.merge(df_2020, on="Name")
```

To prevent bugs like this, we can use the `validate=` option.

```python
df_1920.merge(df_2020, on="Name",
              validate="one_to_one")
```
