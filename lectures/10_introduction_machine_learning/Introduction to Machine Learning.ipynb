```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv("https://dlsun.github.io/pods/data/bordeaux.csv",
                 index_col="year")
df
```

```python
# Filter the dataframes first to align the indices properly
df_train = df[~df["price"].isnull()].reset_index()
df_test = df[df["price"].isnull()].reset_index()
```

```python
px.scatter(df_train,
           x="win", y="summer", color="price", hover_data=['year'])
```

```python
fig1 = px.scatter(df_train,
                  x="win", y="summer", color="price",
                  hover_data=['year']) # Passes the index as 'year'

fig2 = px.scatter(df_test,
                  x="win", y="summer", symbol_sequence=["circle-open"],
                  hover_data=['year'])

fig = go.Figure(data=fig1.data + fig2.data, layout=fig1.layout)
fig.show()
```

```python
import plotly.graph_objects as go

fig1 = px.scatter(df[~df["price"].isnull()],
                  x="win", y="summer", color="price")
fig2 = px.scatter(df[df["price"].isnull()],
                  x="win", y="summer", symbol_sequence=["circle-open"])

go.Figure(data=fig1.data + fig2.data, layout=fig1.layout)
```
