# Encoding Categorical Variables as Quantitative

```python
import pandas as pd
df = pd.read_table("https://datasci112.stanford.edu/data/housing.tsv")
df
```

## Review

Last class, we learned how to calculate the distance between two observations.

```python
features = ["Gr Liv Area", "Bedroom AbvGr", "Full Bath", "Half Bath"]
df[features]
```

```python
import numpy as np
np.sqrt(
    ((df.loc[1707, features] - df.loc[290, features]) ** 2).sum()
)
```

Notice that all of the features were quantitative. What happens if we have some categorical features as well?

```python
features = ["Gr Liv Area", "House Style", "Bedroom AbvGr",
            "Full Bath", "Half Bath", "Neighborhood"]
df[features]
```

We have to convert the categorical variables to quantitative variables.

## Encoding Categorical Features

The standard way to do this is **dummy encoding** or **one-hot encoding**.

```python
pd.get_dummies(df[["House Style"]])
```

You can dummy encode multiple categorical variables at once.

```python
pd.get_dummies(df[["House Style", "Neighborhood"]])
```

If you pass a mix of quantitative variables and categorical variables to `pd.get_dummies()`, it will dummy encode the categorical variables and leave the quantitative variables alone.

```python
pd.get_dummies(df[features])
```

## Column Transformations in Scikit-Learn

```python
from sklearn.preprocessing import OneHotEncoder

# declare the encoder
enc = OneHotEncoder()

# fit the encoder to data
enc.fit(df[["House Style"]])

# transform the data
enc.transform(df[["House Style"]])
```

What is a sparse matrix? How do we see the matrix?

```python
from sklearn.preprocessing import OneHotEncoder

# declare the encoder
enc = OneHotEncoder(sparse_output=False)

# fit the encoder to data
enc.fit(df[["House Style"]])

# transform the data
enc.transform(df[["House Style"]])
```

What if we have a mix of quantitative and categorical variables and only want to dummy encode the categorical ones?

We make a `ColumnTransformer`.


```python
from sklearn.compose import make_column_transformer

transformer = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ["House Style",
                                          "Neighborhood"]),
    remainder="passthrough")
transformer.fit(df[features])
transformer.transform(df[features])
```

Scikit-Learn provides a nice visualization for `ColumnTransformer`.

```python
transformer
```

We can mix scalers and encoders with `ColumnTransformer`!

```python
from sklearn.preprocessing import StandardScaler

transformer = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ["House Style",
                                          "Neighborhood"]),
    (StandardScaler(), ["Gr Liv Area"]),
    remainder="passthrough")
transformer.fit(df[features])
transformer.transform(df[features])
```

Let's visualize this transformer as well.

```python
transformer
```
