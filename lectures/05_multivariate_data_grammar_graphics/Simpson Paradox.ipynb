# Simpson's paradox

```python
import pandas as pd
import matplotlib.pyplot as plt # UI for creating plots (not from pandas)
import plotly.express as px
penguin_df = pd.read_csv("https://datasci112.stanford.edu/data/penguins.csv")
penguin_df
```

Recall the correlations between bill length and depth is negative:

```python
penguin_df.plot.scatter(x="bill_length_mm", y="bill_depth_mm")
```

```python
penguin_df[["bill_length_mm", "bill_depth_mm"]].corr()
```

## Within species

Within each species it is correlated (likely bigger penguins of given species have longer and deeper bills).

```python
px.scatter(penguin_df, x="bill_length_mm", y="bill_depth_mm", color="species")
```

The plot suggests that the correlation between bill length and depth is generally positive within each species.

What is the correlation between bill length and bill depth, conditional on species?

```python
penguin_df.groupby("species")[["bill_length_mm", "bill_depth_mm"]].corr()
```

# Simpson's paradox

We see that in the the aggregate the correlation is negative but 
conditional on species it is generally positive.

Such changes in relationship between two variables when conditioning (or marginalizing) 
on a third is referred to as *Simpson's paradox*.

# Departure delay

We actually saw this (without noting it explicitly)
in the departure delay data.

```python
delay_df = pd.read_csv("https://datasci112.stanford.edu/data/flights_nyc_20131116.csv")
delay_df
```

Let's look at the delay grouped  by `carrier` and then `[carrier, origin]`:

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5)) # make a 1x2 grid of axes
delay_df.groupby("carrier")["dep_delay"].mean().plot.bar(ax=axes[0])
delay_df.groupby(["carrier", "origin"])["dep_delay"].mean().unstack().plot.bar(ax=axes[1])
axes[0].set_ylim(axes[1].get_ylim())
```

Let's focus on Delta (`DL`) and United `(UA)`

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5)) # make a 1x2 grid of axes
focus_df = delay_df[delay_df['carrier'].isin(['UA', 'DL'])]
focus_df.groupby("carrier")["dep_delay"].mean().plot.bar(ax=axes[0])
focus_df.groupby(["carrier", "origin"])["dep_delay"].mean().unstack().plot.bar(ax=axes[1])
axes[0].set_ylim(axes[1].get_ylim())
```

## How is this possible?

- If `UA` has a large majority of its flights at `EWR` then its overall delay would be close to its `EWR` delay (which it seems to be)

- If `DL` has relatively flights from `EWR` then that large delay at `EWR` has small effect on the overall mean.
