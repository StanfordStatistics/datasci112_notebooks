The `DataFrame` object is not native to Python. It is provided by a [Python library called `pandas`](https://pandas.pydata.org/).

```python
import pandas as pd
```

The Titanic data set is stored in a CSV file available online at [https://datasci112.stanford.edu/data/titanic.csv](https://https://datasci112.stanford.edu/data/titanic.csv).


```python
df = pd.read_csv("https://datasci112.stanford.edu/data/titanic.csv")
df
```

We can grab an individual row...

```python
df.loc[2]
```

...or an individual column.

```python
df["pclass"]
```

Remember that **pclass** is a categorical variable.

How do we _summarize_ a categorical variable?



```python
df["pclass"].value_counts()
```

How do we _visualize_ a categorical variable?

```python
pclass_counts = df["pclass"].value_counts()
pclass_counts.plot.bar()
```

Why are the classes out of order? We can fix this by sorting the counts by the index.

```python
df["pclass"].value_counts().sort_index()
```

```python
pclass_counts = df["pclass"].value_counts().sort_index()
pclass_counts.plot.bar()
```
