## Question 4: Putting It All Together

Let's build a single model that includes everything:

- quantitative variables: age, height
- categorical variables: diet, drinks, drugs, smokes
- textual data: aboutme.

To do this, we need to use a `ColumnTransformer` to dummy encode the categorical variables, while vectorizing the textual data. Take a look at the [documentation](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html) for `ColumnTransformer` and see how it's used below. What would happen if we didn't specify the `remainder` argument?

To put all the variables on the same scale, we apply a scaling _after_ transforming all the variables. To do transformations in series, we use a `Pipeline` using `make_pipeline`. The code is provided below.

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

# create a ColumnTransformer to handle categorical and text variables
transformer = make_column_transformer(
    (OneHotEncoder(), []),
    (TfidfVectorizer(), ""),
    remainder="passthrough"  # for the quantitative variables
)

# create a Pipeline that scales all the variables after transforming
pipeline = make_pipeline(
    transformer,
    StandardScaler()
)

# print out the Pipeline
pipeline
```
