## Doing the Compression

Now, let's use clustering to compress this image! This technique is known as **vector quantization**.

Here's the idea:

- Right now, we are storing the colors (RGB values) of the 16384 pixels separately.
- We could save space if we reduced the image to just 16 colors.

Clustering is used to find these 16 colors. Once we find the clusters, each point is (hopefully) near the centroid of its cluster, so we can replace the color of each pixel by the color of its cluster centroid.
