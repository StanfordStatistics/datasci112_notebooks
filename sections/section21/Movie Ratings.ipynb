# Movie Ratings

[MovieLens](http://www.movielens.org/) is a website where users can submit ratings for movies that they watch and receive recommendations for movies. The collected data is made [publicly available for research](http://grouplens.org/datasets/movielens/).

We will be working with a data set of 1 million user ratings of movies. There are more recent data sets with more ratings, but these data sets do not contain demographic information about the users, which is perhaps the most interesting aspect of the data.

The MovieLens data is available at http://dlsun.github.io/pods/data/ml-1m/. The data consists of three files:

- `movies.dat`, which contains information about the movies, such as their title and their genre
- `ratings.dat`, which contains the 1 million ratings
- `users.dat`, which contains information about the users, such as their gender and their occupation

There is also a [README file](http://dlsun.github.io/pods/data/ml-1m/README.txt). You should skim this README file before starting the assignment. In particular, the data files do not have a header row, so you will need this README file to figure out what each column represents.

The code below reads in the `ratings.dat` dataset.

```python
import pandas as pd

data_dir = "https://dlsun.github.io/pods/data/ml-1m/"
columns = "UserID::MovieID::Rating::Timestamp".split("::")
ratings = pd.read_table(data_dir + "/ratings.dat",
                        sep="::", engine="python", encoding="ISO-8859-1",
                        header=None, names=columns)
ratings
```

```python
columns = "MovieID::Title::Genres".split("::")
movies = pd.read_table(data_dir + "/movies.dat",
                        sep="::", engine="python", encoding="ISO-8859-1",
                        header=None, names=columns)
movies
```

```python
columns = "UserID::Gender::Age::Occupation::Zip-code".split("::")

users = pd.read_table(data_dir + "/users.dat",
                        sep="::", engine="python", encoding="ISO-8859-1",
                        header=None, names=columns)
users
```
